```python
#Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the dataset
dataset = pd.read_csv("healthy_lifestyle_city_2021.csv")
dataset
```

## Check if the dataset has missing values and choose a way to correct them. Motivate your choice.

```python
#Check for standard missing values
dataset.isnull().sum().sum()
```

```python
#Making a list of non-standard missing values
missing_values = ["-"]
dataset = pd.read_csv("/Users/lorenzobaglini/Desktop/Università/DATA SCIENCE AND MANAGEMENT/Data visualization/individual homework/healthy_lifestyle_city_2021.csv", na_values = missing_values)

#Total number of missing values
dataset.isnull().sum().sum()
```

```python
#Total missing values for each feature
dataset.isnull().sum()
```

```python
#Replace missing values with average value of attributes, for each feature

#Sunshine hours
sunshine_hours_mean = dataset["Sunshine hours(City)"].mean()
dataset["Sunshine hours(City)"].fillna(sunshine_hours_mean, inplace=True)

#Pollution
pollution_mean = dataset["Pollution(Index score) (City)"].mean()
dataset["Pollution(Index score) (City)"].fillna(pollution_mean, inplace=True)

#Hours worked
hours_worked_mean = dataset["Annual avg. hours worked"].mean()
dataset["Annual avg. hours worked"].fillna(hours_worked_mean, inplace=True)

dataset
```

I chose to replace the missing values in the dataset with the average values of the reference variables, to avoid altering the rest of the data.
Among other alternatives, I could have managed the missing values by inserting, in their place, constants or the value of the median, but I preferred to insert the average values to avoid distortions.
Other possibilities would have been to perform a regression to obtain those specific values, or to insert them manually.

## Plot a bar chart describing the Happiness levels for each city. The bars should be colored in gray except for the bars corresponding to Italian cities which should be colored in dark blue.

```python
#Plot the bar chart related to the happiness levels for each city

colors = ["darkblue" if x == "Milan" else "gray" for x in dataset["City"]]

plt.figure(figsize = (25,10))
plt.bar(dataset["City"],dataset["Happiness levels(Country)"], color = colors)
plt.title("Happiness level for each city")
plt.xlabel("City")
plt.xticks(rotation=45)
plt.ylabel("Happiness levels")

plt.show
```

## Plot the correlation between any two pair of variables as a heatmap with a custom palette.

```python
#Some variables' values aren't considered as numeric value, so I have to convert them

#Cost of a bottle of water
dataset['Cost of a bottle of water(City)'] = dataset['Cost of a bottle of water(City)'].apply(lambda x: x.replace('£', '').replace(',', '')
                                if isinstance(x, str) else x).astype(float)

#Cost of a monthly gym membership
dataset['Cost of a monthly gym membership(City)'] = dataset['Cost of a monthly gym membership(City)'].apply(lambda x: x.replace('£', '').replace(',', '')
                                if isinstance(x, str) else x).astype(float)

#Obesity level
dataset['Obesity levels(Country)'] = dataset['Obesity levels(Country)'].apply(lambda x: x.replace('%', '').replace(',', '')
                                if isinstance(x, str) else x).astype(float)

#Creating and customizing the heatmap
color = sns.color_palette("vlag", as_cmap=True)
sns.heatmap(dataset.corr(),vmin=-1, vmax=1, annot=True, fmt='2.2f', annot_kws={"fontsize":8}, cmap=color)
```

 ### About the result

As can be seen from the heat map, the strongest correlation is between the variables regarding the cost of a bottle of water and the level of happiness of the population. Despite this high value (0.81), it does not appear that the two variables in question can present a strong real correlation.
Therefore, it makes sense to consider strongly correlated variables, those that reflect life expectancy and the level of happiness of people. These two variables have a correlation coefficient of 0.72.
Finally, among the strongly correlated variables, but in a negative sense, the inverse link between people's level of happiness and the degree of pollution in a specific city can be reported.

## Divide the cities into 6 groups based on the first letter of their name (A-D, E-H, I-L, M-P, Q-U, V-Z). 
## Plot a bar chart for the “Number of take out places” variable with six bins (A-D, E-H, I-L, M-P, Q-U, V-Z). 
## Each bin contains the sum of the values of “Number of take out places” for the cities that fall in that bin.

```python
#Sort the dataset by Cities' first letter
data = dataset.sort_values("City")
data
```

```python
#Divide the cities into 6 groups based on the first letter of their name

#A-D cities
df1 = data.loc[:27, ["City", "Number of take out places(City)"]]
#E-H cities
df2 = data.loc[27:30, ["City", "Number of take out places(City)"]]
#I-L cities
df3 = data.loc[30:24, ["City", "Number of take out places(City)"]]
#M-P cities
df4 = data.loc[24:34, ["City", "Number of take out places(City)"]]
#Q-U cities
df5 = data.loc[34:14, ["City", "Number of take out places(City)"]]
#V-Z cities
df6 = data.loc[14:36, ["City", "Number of take out places(City)"]]
```

```python
#Compute the total number of take out places for each group

#A-D cities
dftot1 = df1.sum()
#E-H cities
dftot2 = df2.sum()
#I-L cities
dftot3 = df3.sum()
#M-P cities
dftot4 = df4.sum()
#Q-U cities
dftot5 = df5.sum()
#V-Z cities
dftot6 = df6.sum()
```

```python
#Create a new dataset with the created groups of cities and their relative sum of take out places

#As a first step, create a dictionary
dic = {"Group": ["A-D", "E-H", "I-L", "M-P", "Q-U", "V-Z"], "Sum": [dftot1["Number of take out places(City)"], dftot2["Number of take out places(City)"], dftot3["Number of take out places(City)"], dftot4["Number of take out places(City)"], dftot5["Number of take out places(City)"], dftot6["Number of take out places(City)"]]}

#Then, generate the dataframe
df = pd.DataFrame(dic)
df
```

```python
#Plot a bar chart for the “Number of take out places” variable with six bins.
#Each bin contains the sum of the values of “Number of take out places” for the cities that fall in that bin.
plt.figure(figsize = (25,10))
plt.bar(df["Group"],df["Sum"], color = "orange")
plt.title("Number of take out places for each group of cities", size = 35)
plt.xlabel("Group", size = 30)
plt.xticks(rotation=45, size=20)
plt.ylabel("Take out places", size = 30)
plt.yticks(size = 20)

plt.show
```
